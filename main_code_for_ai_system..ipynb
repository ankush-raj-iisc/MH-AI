```python
# import os

# document = []
# document_count = 0

# for file in os.listdir(docs_folder_path):
#     if file.endswith(".pdf"):
#         pdf_path = docs_folder_path + '/' + file
#         loader = PyPDFLoader(pdf_path)
#         document.extend(loader.load())
#         document_count += 1
#     elif file.endswith('.docx') or file.endswith('.doc'):
#         doc_path = docs_folder_path + '/' + file
#         loader = Docx2txtLoader(doc_path)
#         document.extend(loader.load())
#         document_count += 1
#     elif file.endswith('.txt'):
#         text_path = docs_folder_path + '/' + file
#         loader = TextLoader(text_path)
#         document.extend(loader.load())
#         document_count += 1

# print(f"Processed {document_count} documents.")

import os
import threading
from queue import Queue

# Define a thread-safe queue
queue = Queue()

# Define a lock for thread-safe prints
print_lock = threading.Lock()

def process_file():
    while True:
        # Get a file from the queue
        file = queue.get()

        try:
            if file.endswith(".pdf"):
                pdf_path = docs_folder_path + '/' + file
                loader = PyPDFLoader(pdf_path)
                document.extend(loader.load())
            elif file.endswith('.docx') or file.endswith('.doc'):
                doc_path = docs_folder_path + '/' + file
                loader = Docx2txtLoader(doc_path)
                document.extend(loader.load())
            elif file.endswith('.txt'):
                text_path = docs_folder_path + '/' + file
                loader = TextLoader(text_path)
                document.extend(loader.load())

            with print_lock:
                print(f"Processed {file}")
        except Exception as e:
            with print_lock:
                print(f"Error processing {file}: {e}")
        finally:
            # Signal that the task is done
            queue.task_done()

document = []

# Create 10 worker threads
for i in range(10):
    worker = threading.Thread(target=process_file, daemon=True)
    worker.start()

# Put the files in the queue
for file in os.listdir(docs_folder_path):
    queue.put(file)

# Wait for all the tasks in the queue to be processed
queue.join()

print(f"Processed {len(document)} documents.")
```

```python
# Initialize an empty list to store the answers
answers = []

# Open the text file in append mode
with open("output.txt", "a") as file:
    # Iterate over the first 2 questions in the DataFrame
    for i in range(2):
        question = df["Questions"].iloc[i]
        # Run the question through the QA model
        result = qa(question)
        # Append the answer to the list
        answers.append(result["answer"])
        # Write the question, answer, and context to the file
        file.write(f"Question: {question}\n")
        file.write(f"Answer: {result['answer']}\n")
        file.write(f"Context: {result['context']}\n")
        file.write("\n")

# Add the answers as a new column in the DataFrame
df["answer"] = pd.Series(answers)
```

```python
# Initialize the retriever
retriever = vectordb.as_retriever(search_kwargs={'k':10})

# Iterate over the first 2 questions in the DataFrame
for i in range(2):
    question = df["Questions"].iloc[i]
    # Use the retriever to fetch the relevant context
    context = await retriever.aget_relevant_documents(query=question)
    # Now you have the relevant context for the question
    print(f"Question: {question}\nContext: {context}\n")
```
