### __The below cell can be used to extract text from any pdf using the PyPDF library__

```python

import PyPDF2

# Open the PDF file in read binary mode
with open('C://Users//dixit//Downloads//testPdfDelete.PDF', 'rb') as pdf_file:
    # Create a PyPDF2 PDF reader object
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    
    # Extract text content from each page in the PDF file
    text_content = ''
    for page_num in range(len(pdf_reader.pages)):
        page_obj = pdf_reader.pages[page_num]
        text_content += page_obj.extract_text()
    
    # Print the extracted text content
    print(text_content)
```

### __The below cell will be used to scrape the one page of the website__

```python
import time
import requests
import PyPDF2
import os
import tempfile
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver

path='.\chromedriver.exe'
# set up the webdriver
driver = webdriver.Chrome(path)
driver.maximize_window()

# navigate to the webpage
driver.get("https://sat.gov.in/scripts/search.asp")

# find the element using xpath and click on it
elm = 4
while True:
    try:
        element = driver.find_element(By.XPATH,f'//*[@id="form1"]/table/tbody/tr[{elm}]/td[4]/p/a')
        element.click()

        # wait for the pdf window to load
        time.sleep(5)

        # switch to the pdf window
        pdf_window = driver.window_handles[-1]
        driver.switch_to.window(pdf_window)
        current_url = driver.current_url
        print(current_url)
        
        # download the pdf file to a local directory
        response = requests.get(current_url)
        file = tempfile.NamedTemporaryFile(delete=False)
        with open(file.name, 'wb') as f:
            f.write(response.content)

        # extract the text from the pdf file
        with open(file.name, 'rb') as f:
            pdf_file = PyPDF2.PdfReader(f)
            text = ''
            for page in range(len(pdf_file.pages)):
                page_text = pdf_file.pages[page].extract_text()
                text += page_text

        # delete the temporary file
        print(text)
        file.close()
        os.remove(file.name)

        

        print(f"Interacted with the pdf window {elm} and now closing the pdf window {elm} ...")
        driver.close()
        elm += 2
        driver.switch_to.window(driver.window_handles[0])
    except NoSuchElementException:
        print("No such element found")
        break

print(f"Job done")
driver.quit()
```

**the same as above code but will create dict instead of printing the whole text to the console**

```python
import time
import requests
import PyPDF2
import os
import tempfile
import json
import re
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver

path='.\chromedriver.exe'
# set up the webdriver
driver = webdriver.Chrome(path)
driver.maximize_window()


# navigate to the webpage
driver.get("https://sat.gov.in/scripts/search.asp")

# create an empty dictionary to store the PDF text
pdf_text_dict = {}

# find the element using xpath and click on it
elm = 4
while True:
    try:
        element = driver.find_element(By.XPATH,f'//*[@id="form1"]/table/tbody/tr[{elm}]/td[4]/p/a')
        website_name = element.text
        element.click()

        # wait for the pdf window to load
        time.sleep(5)

        # switch to the pdf window
        pdf_window = driver.window_handles[-1]
        driver.switch_to.window(pdf_window)
        current_url = driver.current_url
        
        # download the pdf file to a local directory
        response = requests.get(current_url)
        file = tempfile.NamedTemporaryFile(delete=False)
        with open(file.name, 'wb') as f:
            f.write(response.content)

        # extract the text from the pdf file
        with open(file.name, 'rb') as f:
            pdf_file = PyPDF2.PdfReader(f)
            text = ''
            for page in range(len(pdf_file.pages)):
                page_text = pdf_file.pages[page].extract_text()
                text += page_text

        # extract index and case date from the text using regex
        index = re.findall(r'Index No.\s*:\s*([^\n]+)', text)
        case_date = re.findall(r'Case Date\s*:\s*([^\n]+)', text)

        # create a nested dictionary to store the index and case date with the text
        data_dict = {"index": index, "case date": case_date, "text": text}

        # add the nested dictionary to the main dictionary with the website as the key
        pdf_text_dict[website_name] = data_dict

        # delete the temporary file
        file.close()
        os.remove(file.name)

        print(f"Interacted with the pdf window {elm} and now closing the pdf window {elm} ...")
        driver.close()
        elm += 2
        driver.switch_to.window(driver.window_handles[0])
    except NoSuchElementException:
        print("No such element found")
        break

# save the dictionary to a JSON file
with open('pdf_text.json', 'w') as f:
    json.dump(pdf_text_dict, f)

print(f"Job done")
driver.quit()
```
