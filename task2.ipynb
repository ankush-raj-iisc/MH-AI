```python
import io
import PyPDF2
from selenium import webdriver
import time

# create a webdriver instance and navigate to the PDF URL in Chrome
driver = webdriver.Chrome()
driver.get('https://example.com/your/file.pdf')

# wait for the PDF to load
time.sleep(5)

# get the PDF content as bytes
pdf_content = driver.execute_script("return document.documentElement.outerHTML").encode('utf-8')

# create a PyPDF2 PdfFileReader object from the PDF content
pdf_reader = PyPDF2.PdfFileReader(io.BytesIO(pdf_content))

# extract the text from each page of the PDF
text = ''
for page_num in range(pdf_reader.numPages):
    page = pdf_reader.getPage(page_num)
    text += page.extractText()

# print the extracted text
print(text)

# close the webdriver instance
driver.quit()
```

```python
import time
import io
import urllib.request
from pikepdf import Pdf
from selenium import webdriver
from selenium.webdriver.common.by import By

path='.\chromedriver.exe'
# set up the webdriver
driver = webdriver.Chrome(path)
driver.maximize_window()

# navigate to the webpage
driver.get(website)

# find the element using xpath and click on it
for elm in range(48,1416,2):
    element = driver.find_element(By.XPATH,f'//*[@id="form1"]/table/tbody/tr[{elm}]/td[4]/p/a')
    element.click()

    # wait for the pdf window to load
    time.sleep(5)

    # switch to the pdf window
    pdf_window = driver.window_handles[-1]
    driver.switch_to.window(pdf_window)
    print(f"Interacted with the pdf window {elm}")

    # extract the text from the pdf
    pdf_text = ""
    with urllib.request.urlopen(driver.current_url) as response:
        with io.BytesIO(response.read()) as f:
            with Pdf.open(f) as pdf:
                for page in pdf.pages:
                    pdf_text += page.extract_text

    # switch back to the previous window
    driver.close()
    driver.switch_to.window(driver.window_handles[-1])
    print(pdf_text)
    
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import PyPDF2
import os
import time

# Set up Chrome options to download files automatically
chrome_options = Options()
chrome_options.add_experimental_option('prefs', {
    'download.default_directory': os.getcwd(),
    'download.prompt_for_download': False,
    'download.directory_upgrade': True,
    'plugins.always_open_pdf_externally': True
})

# Create a new instance of Chrome webdriver
service = Service('/chromedriver.exe')
driver = webdriver.Chrome(service=service, options=chrome_options)

# Navigate to the website with the PDF files
driver.get('https://sat.gov.in/scripts/search.asp')

# Find all the file links and click on each one
file_links = driver.find_elements(By.XPATH ,'//a[contains(@href, ".pdf")]')
print(f'Found {len(file_links)} PDF links on the page')

# Find all the links on the page
links = driver.find_elements(By.XPATH, '//a')
print(type(links))
print(links)
links = [file for file in links if file.endswith('.PDF')]
# Print out the attributes of each link
# for link in links:
#     href = link.get_attribute('href')
#     text = link.text
#     print(f'Link: {text} ({href})')


for i, link in enumerate(links):
    link.click()
    print("here in the loop ")

    # Wait for the file to download (you may need to adjust this time depending on the size of the file)
    time.sleep(10)

    # Get the filename of the downloaded file
    filename = max([os.path.join(os.getcwd(), f) for f in os.listdir(os.getcwd())], key=os.path.getctime)

    # Open the PDF file and read its contents
    with open(filename, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        contents = ''
        for i in range(reader.getNumPages()):
            contents += reader.getPage(i).extract_text()

    # Print the contents of the PDF file
    print(f'Contents of file {i+1}:')
    print(contents)

    # Delete the PDF file
    os.remove(filename)

# Quit the webdriver
driver.quit()
```
