```python
# Define the file path
file_path = 'raw.txt'  # Replace with the path to your text file

# Initialize an empty list to store the URLs
url_list = []

# Open the file in read mode with the appropriate encoding
with open(file_path, 'r', encoding='utf-8') as file:
    # Iterate through each line in the file
    for line in file:
        # Remove leading and trailing whitespace, if any
        line = line.strip()
        
        
        # Check if the line is a valid URL (you may want to enhance URL validation)
        if line.startswith('[') or line.startswith('http://'):
            url_list.append(line)

# Print the list of extracted URLs
print(url_list)
```

```python
def remove_chars_from_list(strings):
    # Initialize an empty list to store the modified strings
    modified_strings = []
    
    # Iterate through the input list of strings
    for string in strings:
        # Check if the string has at least 4 characters (to avoid removing too many characters)
        if len(string) >= 4:
            # Remove the first two and last two characters
            modified_string = string[2:-2]
            # Append the modified string to the result list
            modified_strings.append(modified_string)
        else:
            # If the string has fewer than 4 characters, leave it unchanged
            modified_strings.append(string)
    
    return modified_strings

# Example usage with your url_list

modified_url_list = remove_chars_from_list(url_list)

print(modified_url_list)
```
