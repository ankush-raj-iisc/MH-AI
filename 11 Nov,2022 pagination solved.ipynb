```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import time
import os

# Configure Chrome WebDriver to run in headless mode
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # Add this line to run in headless mode
chrome_options.add_argument('--disable-gpu')

# Define the path to the Chrome WebDriver
chromedriver_path = '.\chromedriver.exe'

# Function to save data to an Excel file
def save_to_excel(data, file_name):
    df = pd.DataFrame(data, columns=['Text'])
    df.to_excel(file_name, index=False)

# Check if the Excel file already exists and load data
excel_file = 'output_data.xlsx'

if os.path.exists(excel_file):
    existing_data = pd.read_excel(excel_file)
    res = existing_data['Text'].tolist()
else:
    res = []

# Initialize the web driver and navigate to the website
driver = webdriver.Chrome(chromedriver_path)
driver.maximize_window()
driver.implicitly_wait(5)
#website = 'https://example.com'  # Replace with your target website
driver.get(website)
driver.implicitly_wait(5)

# Scraping loop
page = 1
while page < 2:  # Change the number of pages to scrape as needed

    element = driver.find_element(By.XPATH,"(//a[@class='active'])[1]") 
    actionChains = ActionChains(driver)
    actionChains.context_click(element).perform()
    
    for i in range(1, 500):  # Change the range as needed

        driver.find_element(By.XPATH, f"(//a[@class='points'])[{i}]").click()
        oldURL = driver.window_handles[0]
        newURl = driver.window_handles[1]
        driver.switch_to.window(newURl)
        url = driver.current_url

        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'})
        content = BeautifulSoup(response.text, 'html.parser')

        if 'iframe' in content.prettify():        
            time.sleep(5)
            driver.switch_to.frame(driver.find_element(By.XPATH, "//iframe"))
            driver.implicitly_wait(5)

            num_pages = driver.find_elements(By.XPATH, "//div[@class='page']")
            p = []
            page_num = len(num_pages)

            if page_num == 2:
                for i in range(1, page_num + 1):
                    driver.find_element(By.XPATH,'//input[@id="pageNumber"]').send_keys(Keys.CONTROL + "a")
                    driver.find_element(By.XPATH,'//input[@id="pageNumber"]').send_keys(i)
                    driver.find_element(By.XPATH,'//input[@id="pageNumber"]').send_keys(Keys.ENTER)

                    programs = driver.find_elements(By.XPATH, f"(//div[@data-page-number='{i}'])[2]")
                    driver.implicitly_wait(5)
                    for program in programs:
                        a = program.text
                        p.append(a)

                p = ' '.join(p)
                res.append(p.replace('\n', ' '))
                driver.close()
                driver.switch_to.window(oldURL)
                url = driver.current_url

            else:
                for i in range(1, page_num + 1):
                    driver.find_element(By.XPATH,'//input[@id="pageNumber"]').send_keys(Keys.CONTROL + "a")
                    driver.find_element(By.XPATH,'//input[@id="pageNumber"]').send_keys(i)
                    driver.find_element(By.XPATH,'//input[@id="pageNumber"]').send_keys(Keys.ENTER)

                    programs = driver.find_elements(By.XPATH, f"(//div[@data-page-number='{i}'])[2]")
                    driver.implicitly_wait(5)
                    for program in programs:
                        a = program.text
                        p.append(a)

                p = ' '.join(p)
                res.append(p.replace('\n', ' '))
                driver.close()
                driver.switch_to.window(oldURL)
                url = driver.current_url
        else:
            driver.close()
            driver.switch_to.window(oldURL)
            url = driver.current_url

    print(url)
    driver.find_element(By.XPATH, "(//i[@class='fa'])[2]").click()
    page = page + 1

    # Save data to Excel at each step
    save_to_excel(res, excel_file)

# Close the web driver
driver.quit()

# Save the final data to Excel
save_to_excel(res, excel_file)
```

```python
import requests
from bs4 import BeautifulSoup

def list_urls_on_webpage(url):
    try:
        # Send a GET request to the URL and get the HTML content
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find all anchor tags (a) with 'href' attribute and extract the URLs
            links = [a['href'] for a in soup.find_all('a', href=True)]
            
            return links
        else:
            print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None

# # Example usage:
# website_url = 'YOUR_WEBSITE_URL'
# urls = list_urls_on_webpage(website_url)
# if urls:
#     for url in urls:
#         print(url)
```

```python
def extract_substring(text):
    # Convert the input text to lowercase for case-insensitive matching
    text = text.lower()
    
    # Find the starting index of "background" in the lowercase text
    start_index = text.find("background")
    
    if start_index != -1:
        # Find the starting index of "issue 2" after the "background"
        end_index = text.find("issue no. iii", start_index)
        
        if end_index != -1:
            # Extract the substring from "background" to "issue 2"
            return text[start_index:end_index + len("issue no .3")]
    
    # Return None if no match is found
    return None
```

```python
import pandas as pd

# Define the extract_substring function as shown earlier

# Read the Excel file into a pandas DataFrame
excel_file = "output_data.xlsx"  # Replace with the path to your Excel file
df = pd.read_excel(excel_file)

# Apply the extract_substring function to the "text" column and store the results in a new column

df['extracted_text'] = df['Text'].apply(extract_substring)

# Save the modified DataFrame back to a new Excel file if needed
output_excel_file = "output_excel_file.xlsx"  # Replace with your desired output file path
df.to_excel(output_excel_file, index=False)

# Optionally, you can print the DataFrame to see the results
print(df)
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

def extract_urls_with_selenium(url, num_pages=5):
    try:
        chrome_driver_path = './chromedriver.exe'

        # Create a Selenium WebDriver instance (this will open a browser window)
        driver = webdriver.Chrome(executable_path=chrome_driver_path)

        # Navigate to the initial webpage
        driver.get(url)

        # Initialize an empty list to store URLs
        all_links = []
        

        # Iterate through the pages and click the JavaScript buttons
        for page_num in range(400, num_pages + 1):
            script = f"searchFormNewsList('n', '{page_num}');"
            driver.execute_script(script)
            time.sleep(2)  # Wait for the page to load (adjust as needed)

            # Get the page source after clicking the button
            page_source = driver.page_source

            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(page_source, 'html.parser')

            # Find all anchor tags (a) with 'href' attribute and filter for the desired format
            links = [a['href'] for a in soup.find_all('a', href=True) if 'https://www.sebi.gov.in/enforcement/orders/' in a['href']]

            # Extend the list of URLs
            all_links.extend(links)
            print(f"Interacted with page {page_num}")

        # Create a Pandas DataFrame from the list of URLs
        df = pd.DataFrame({'URLs': all_links})

        return df
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None
    finally:
        # Close the browser when done
        driver.quit()

# Usage:
website_url = 'https://www.sebi.gov.in/sebiweb/home/HomeAction.do?doListing=yes&sid=2&ssid=9&smid=6'
df = extract_urls_with_selenium(website_url, num_pages=446)  # You can specify the number of pages to scrape
if df is not None:
    print(df)
```
