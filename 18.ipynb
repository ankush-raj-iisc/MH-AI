```python
import pandas as pd

def column_to_list(data_frame, column_name):
    """
    Converts a column from a DataFrame into a list.

    Args:
        data_frame (pd.DataFrame): The DataFrame containing the column.
        column_name (str): The name of the column to convert.

    Returns:
        list: A list containing the values from the specified column.
    """
    if column_name in data_frame.columns:
        return data_frame[column_name].tolist()
    else:
        raise ValueError(f"Column '{column_name}' not found in the DataFrame.")

# Example usage:
# Assuming you have a DataFrame 'df' and want to convert the 'column_name' column into a list:
my_list = column_to_list(df, 'PDFUrls')
```
